```python
import random

def generate_matrix_with_nonzero_integer_determinant(n, max_coef):
    # Start with an identity matrix of size n
    matrix = [[1 if i == j else 0 for j in range(n)] for i in range(n)]

    # Perform a sequence of row operations
    for i in range(n):
        # Choose a random row to swap with row i
        j = random.randint(i, n-1)
        matrix[i], matrix[j] = matrix[j], matrix[i]

        # Choose a random scalar within the specified range and add it to row i
        scalar = random.randint(-max_coef, max_coef)
        for j in range(n):
            coef = random.randint(-max_coef, max_coef)
            matrix[i][j] += coef * scalar * matrix[i-1][j]

    # The final matrix has a non-zero integer determinant
    return matrix


import numpy as np
M = np.matrix(generate_matrix_with_nonzero_integer_determinant(3, 3))
print(np.linalg.det(M))
print(np.linalg.solve(M, np.zeros(shape=M.shape[0])))
print(M)
```

```python
import random
import numpy as np

def generate_matrix_and_vector(n, maxval_M, maxval_b, maxval_x):
    M = np.zeros((n, n), dtype=int)
    x = np.zeros((n,), dtype=int)
    b = np.zeros((n,), dtype=int)

    for i in range(n):
        x[i] = random.randint(1, maxval_x)

    for i in range(n):
        for j in range(n):
            M[i, j] = random.randint(1, maxval_M)

        # Ensure that the dot product of M and x doesn't exceed maxval_b
        while True:
            b[i] = np.dot(M[i], x)
            if b[i] <= maxval_b:
                break
            else:
                # Update the coefficients of M if the dot product exceeds maxval_b
                for j in range(n):
                    M[i, j] = random.randint(1, maxval_M)

    return M, b, x

n = 3
maxval_M = 10
maxval_b = 20
maxval_x = 4

M, b, x = generate_matrix_and_vector(n, maxval_M, maxval_b, maxval_x)

print("Matrix M:")
print(M)
print("\nVector b:")
print(b)
print("\nSolution Vector x:")
print(x)

print("Verification")
print(M.astype(float)@x.astype(float), b)

```

```python
import numpy as np
import random

def generate_filled_grid(n):
    def is_valid_move(x, y):
        return 0 <= x < n and 0 <= y < n and grid[x][y] == 0

    def count_neighbors(x, y, shape_id):
        count = 0
        for dx, dy in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
            nx, ny = x + dx, y + dy
            if 0 <= nx < n and 0 <= ny < n and grid[nx][ny] == shape_id:
                count += 1
        return count

    def dfs(x, y, shape_id):
        if not is_valid_move(x, y):
            return

        neighbors = count_neighbors(x, y, shape_id)

        if neighbors >= 2:  # Prevent more than 2 connections
            return

        grid[x][y] = shape_id
        directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]
        random.shuffle(directions)

        for dx, dy in directions:
            dfs(x + dx, y + dy, shape_id)

    grid = np.zeros((n, n), dtype=int)
    shape_id = 1

    for x in range(n):
        for y in range(n):
            if grid[x][y] == 0:
                dfs(x, y, shape_id)
                shape_id += 1

    return grid

n = 4
filled_grid = generate_filled_grid(n)
print(filled_grid)
```
